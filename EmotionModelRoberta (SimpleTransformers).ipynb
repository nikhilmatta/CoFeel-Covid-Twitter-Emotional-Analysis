```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('Twitter Covid Data/emotionDataset.csv')
```

```python
dataset.drop(columns='Unnamed: 0',inplace=True,axis=1)
```

```python
dataset = dataset.reindex(columns=['tweet','emotion'])
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
dataset['emotion'] = labelencoder.fit_transform(dataset['emotion'])
```

```python
dataset
```

```python
dataset = dataset.rename(columns={"tweet": "text", "emotion": "labels"})
```

```python
from simpletransformers.classification import ClassificationModel
```

```python
train_size = 0.75
train_dataset=dataset.sample(frac=train_size,random_state=200).reset_index(drop=True)
test_dataset=dataset.drop(train_dataset.index).reset_index(drop=True)
```

```python
print("FULL Dataset: {}".format(dataset.shape))
print("TRAIN Dataset: {}".format(train_dataset.shape))
print("TEST Dataset: {}".format(test_dataset.shape))
```

```python
train_dataset
```

```python
train_args={
    'evaluate_during_training': True,
    'output_dir': 'TwitterEmotionAnalysis/robertaModelV4',
    'num_train_epochs': 10,
    'train_batch_size': 150,
    'eval_batch_size': 100,
    'reprocess_input_data': True,
    'learning_rate': 1e-5
}
```

```python
model=ClassificationModel('roberta', 'roberta-base', use_cuda=True, args=train_args,num_labels=13)
```

```python
train_dataset.dropna(how='any',inplace=True)
test_dataset.dropna(how='any',inplace=True)
```

```python
train_dataset.reset_index(drop=True)
test_dataset.reset_index(drop=True)
```

```python
# !pip uninstall apex
# !cd apex

#!python setup.py install --cuda_ext --cpp_ext
```

```python
model.train_model(train_dataset,eval_df=test_dataset)
```
